Besides the atmosphere, we can use box modeling to evaluate the ocean system as well. Deciding on the number of boxes to include will determine what we can model. # 1-Box Ocean Model

A basic 1-box model of the ocean to model biogeochemical changes typically consists of a river source that introduces a chemical to the ocean and a sediment loss process to remove the chemical from the ocean.

Ocean_1Box.JPG

The [phosphorus cycle](https://geo.libretexts.org/Bookshelves/Geography_(Physical)/BioGeoChemistry_(LibreTexts)/12%3A_The_Global_Cycles_of_Nitrogen_Phosphorus_and_Potassium/12.05%3A_The_global_phosphorus_cycle#:~:text=Phosphates%20%28PO%2043%E2%88%92%29%20are%20sent%20into%20rivers%2C%20lakes%2C,the%20uplifting%20of%20areas%20of%20the%20earth%E2%80%99s%20surface.) is very important to biogeochemists as phosphate ($\text{PO}_4^{-3}$) is a limiting bionutrient to biological production in the ocean. Understanding the controls on phosphate in the ocean help us to better understand how and why changes in past ocean composition occured.

To create our simple model, we will use:

* $V_{oc} = 1.29 \times 10^{18} \ \text{m}^3$
* $V_{riv} = 3.7 \times 10^{13} \ \frac{\text{m}^3}{\text{yr}}$

Where $V_{oc}$ and $V_{riv}$ are the total ocean volume and annual river input volume to the ocean, respectively.

# Ocean 2-Box Model

Our ocean model can be made more interesting by splitting the ocean into a deep ocean and a surface mixed layer.

Ocean_2Box.JPG



The amount of $\text{PO}_4^{-3}$ in the surface ocean is tightly coupled to organic matter (*OM*) production (*uptake*), the total import of $\text{PO}_4^{-3}$ from rivers, and the vertical mixing between the surface mixed layer and deep ocean layer.

In the deep ocean, the amount of $\text{PO}_4^{-3}$ depends on the amount of remineralization of organic matter (*decomposition*) and vertical mixing between the surface and deep ocean.

It is worth noting that the amount lost to transport to sediments (*OM Burial*) is only relavent in keeping track of the total amount of phosphorous in the system. It doesn't directly impact the amount of phosphate in the deep ocean as an increase in the amount buried results in a decreas in the amount of remineralization.

## Q7

A 2-box model will have two coupled differential equations that describe how the concentration in each box changes over time.

Our equations for this situation will be:

\begin{align}
V_{s}\frac{d[C]_s}{dt} &= V_{riv}[C]_{riv} + \nu([C]_d - [C]_s) - \Phi_{uptake} \tag{4} \\
\\
V_{d}\frac{d[C]_d}{dt} &= \Phi_{decomposition} - \nu([C]_d - [C]_s)  \tag{5}
\end{align}

Note that we have ignored the burial term as a loss mechanism for the deep ocean.

Our system of equations has several unknowns that we will need to find:
* $\Phi_{uptake}$, biological uptake of phosphorus in the surface ocean
* $\Phi_{decomposition}$, organic matter decomposition in the deep ocean
* $\nu$, vertical mixing rate between ocean layers
* $[C]_s$, concentration in the mixed layer ocean
* $[C]_d$, concentration in the deep ocean

So we need to either make assumptions or add more information to help us solve the system.

We can use the same $V_{riv}$ and $[C]_{riv}$ as our 1-Box model. For the vertical mixing, we can use a value derived from radiocarbon measurements, $\nu = 1.2 \times 10^{15} \frac{\text{m}^3}{yr}$.

Because phosphate is a biolimiting chemical in the surface ocean, it is reasonable to assume the rate of biological uptake, $\Phi_{uptake}$, is proportional to the concentration of $\text{PO}_4^{-3}$ in the surface waters:

$$
\begin{align}
\Phi_{uptake} &= k_s V_s [C]_s \tag{6}
\end{align}
$$

Where $k_s$ is the rate constant describing the rate at which organic matter is forming in surface waters.

**Assuming we are in steady-state, calculate the value of the rate constant $k_s$ describing the rate at which organic matter is forming in surface waters.**

## Q8

We still need to define, $\Phi_{decomposition}$. It turns out we don't need to know the rate at which organic matter is exported from the surface to the deep ocean because we are only concerning ourselves with the concentration of $\text{PO}_4^{-3}$ in each box, not the total number of phosphorous atoms in each box.

This means we can use conservation of mass to claim that the organic matter that formed in the surface ocean, $\Phi_{uptake}$, eventually sinks to the deep ocean where it will either undergo remineralization, $\Phi_{decomposition}$, or become buried in the sediments, $L$.

This reduces our problem to defining what fraction ($f$) of $\Phi_{uptake}$ is buried and what fraction will undergo remineralization.

To find this, we simply need to start with steady-state conditions and use the face we know $L = f \Phi_{decomposition}$
