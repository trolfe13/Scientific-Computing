$$
ax^2 + bx + c = 0 \tag{1}
$$

has an analytic solution that can be written as:

$$
\begin{align}
x_{1,2} &= \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} \tag{2} \\
x'_{1,2} &= \frac{-2c}{b \pm \sqrt{b^2 - 4ac}} \tag{3}
\end{align}
$$

Note that the "primed" equation ($x'$) is not the derivative, but simply an alternative expression of the more familiar $x$ equation.
