# Data Types

In Python, data types are used to classify the different kinds of values that variables can hold. They determine the operations that can be performed on the data and how the data is stored in memory. Here are some of the built-in data types in Python:

**Numeric Types**:

* `int`: Represents integers (whole numbers) without any decimal points. For example, 10, -5, 0.
* `float`: Represents floating-point numbers (numbers with decimal points). For example, 3.14, -2.5, 0.0.
* `complex`: Represents complex numbers with a real and imaginary part. For example, 2 + 3j.

**Text Type**:

* `str`: Represents strings of characters enclosed in single quotes ('), double quotes ("), or triple quotes (''' or """).

**Boolean Type**:

* `bool`: Represents Boolean values, which can be either True or False. These are often used in conditional statements and logical operations.

**Sequence Types**:

* `list`: An ordered, mutable (changeable) collection of items. Items can be of different data types. For example, [1, 2, 'hello', 3.14].
* `tuple`: An ordered, immutable (unchangeable) collection of items. Items can be of different data types. For example, (1, 2, 'hello', 3.14).
* `range`: Represents a sequence of numbers, often used in loops. For example, range(1, 5) generates the sequence 1, 2, 3, 4.

**Mapping Type**:

* `dict`: Represents a dictionary, which stores key-value pairs. Keys must be immutable, while values can be of any data type. For example, {'name': 'Alice', 'age': 30}.

**Set Types**:

* `set`: An unordered collection of unique items. Items must be immutable. For example, {1, 2, 3}.
* `frozenset`: An immutable version of a set.

**Binary Types**:

* `bytes`: Represents a sequence of bytes.
* `bytearray`: A mutable version of bytes.
* `memoryview`: Allows access to the internal memory of an object without copying.


You can use the `type()` function to check the data type of a variable.

## List vs Tuple and Indexs

It is worth specifically noting the difference between a list and a tuple. A list is mutable (changeable) while a tuple is nonmutalbe (unchangeable). For instance, let's say we have a list of numbers:


```
a = [10, 12, 13, 14]
```

If we want to change the first element (index 0):


```
a[0] = 11  # this is allowed
print(a)   # output: [11, 12, 13, 14]
```

Python will allow us to write over the previous value and save the new value. However, if we used the tuple format instead, Python will produce an error message.


```
b = (10, 12, 13, 14)
b[0] = 11    # this will raise an error!
```

So, why would we be interested in using lists over tuples or vise versa? Obviously if you want to write over previous values, you will want to use a list or if you need the values to remain unchanged, a tuple will be the way to go. Another factor to consier is computation time. A list requires different memory storage than a tuple (tuples are stored in a single memory block while lists are stored in two). Because of this, tuples will generally be computed faster than lists. The code below gives an example of this. In this class, we are not going to focus heavily on computational efficiency, but it is something you may need to consider in the future.

````python
import time
l = list(range(1000000))
t = tuple(range(1000000))
start = time.time()
for i in l:
  pass
print("List iteration time:", time.time() - start)
start = time.time()
for i in t:
  pass
print("Tuple iteration time:", time.time() - start)
````

## NumPy

[NumPy](https://numpy.org/) (Numerical Python) is a fundamental package for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.

NumPy is often going to be used to help us work with datasets that have many rows and columns worth of numbers.

**Key Features of NumPy**

* N-dimensional Arrays: NumPy's primary data structure is the ndarray, which allows for efficient storage and manipulation of large datasets.

* Mathematical Functions: It includes comprehensive mathematical functions, random number generators, linear algebra routines, Fourier transforms, and more.

* Interoperability: NumPy supports a wide range of hardware and computing platforms and integrates well with distributed, GPU, and sparse array libraries.

**Basic Usage**

To use NumPy, you need to import it into your Python code. The standard convention is to import it as np:

`import numpy as np`

**Creating Arrays**

You can create a NumPy array from a Python list:

```
data = np.array([1, 2, 3, 4, 5])
print(data)
print(type(data))
```

**Array Operations**

NumPy allows for a variety of operations on arrays, such as addition, subtraction, multiplication, and division:

```
data = np.array([1, 2])
ones = np.ones(2, dtype=int)
print(data + ones) # Output: [2, 3]
print(data - ones) # Output: [0, 1]
print(data * data) # Output: [1, 4]
print(data / data) # Output: [1.0, 1.0]
```


**Array Attributes**

NumPy arrays have several attributes that provide information about the array:

`ndim`: Number of dimensions

`shape`: Shape of the array

`size`: Total number of elements

`dtype`: Data type of the elements

Example:

```
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a.ndim)  # Output: 2
print(a.shape) # Output: (2, 3)
print(a.size)  # Output: 6
print(a.dtype) # Output: int64
```

**Indexing and Slicing**

You can index and slice NumPy arrays similarly to Python lists:

```
data = np.array([1, 2, 3])
print(data[1])   # Output: 2
print(data[0:2]) # Output: [1, 2]
print(data[1:])  # Output: [2, 3]
print(data[-2:]) # Output: [2, 3]
```

**Reshaping Arrays**

You can reshape arrays without changing their data using the reshape method:

```
a = np.arange(6)
b = a.reshape(3, 2)
print(b)
# Output:
# [[0 1]
# [2 3]
# [4 5]]
```

**Aggregation Functions**

NumPy provides several aggregation functions such as sum, min, max, mean, std, etc.:

```
a = np.array([1, 2, 3, 4])
print(a.sum())  # Output: 10
print(a.min())  # Output: 1
print(a.max())  # Output: 4
print(a.mean()) # Output: 2.5
print(a.std())  # Output: 1.118033988749895
```

# Loops

In Python, loops are constructs that allow us to execute a block of code repeatedly based on a condition. There are two main types of loops in Python: for loops and while loops. Both loops have their own use cases and syntax.

## For Loops

A for loop is used for iterating over a sequence (such as a list, tuple, dictionary, set, or string). It works more like an iterator method found in other object-oriented programming languages. Here is an example of a for loop iterating over a list of fruits:

```
fruits = ["apple", "banana", "cherry"]
for i in fruits:
  print(i)
```

You can also loop through the characters in a string:

```
for i in "banana":
  print(i)
```

### The `range()` Function

The `range()` function is commonly used with for loops to generate a sequence of numbers. For example:

```
for i in range(6):
  print(i)
```

This will print numbers from 0 to 5. You can also specify a start value and an increment value:

```
for i in range(2, 6):
  print(i)

for i in range(2, 30, 3):
  print(i)
```

### Break and Continue Statements

The break statement can be used to exit the loop prematurely:

```
for i in fruits:
  if i == "banana":
    break
  print(i)
```

The continue statement can be used to skip the current iteration and continue with the next:

```
fruits = ["apple", "banana", "cherry"]
for i in fruits:
  if i == "banana":
    continue
  print(i)
```

### Else in For Loop

The else block in a for loop is executed when the loop finishes normally:

```
for i in range(6):
  print(i)
else:
  print("Finally finished!")
```

## While Loops

A while loop is used to execute a block of statements repeatedly until a given condition is satisfied. The syntax is:

```
while condition:
  statement(s)
```

Here is an example of a while loop:

```
count = 0
while count < 3:
  count += 1
  print("Hello")
```

### Else in While Loop

The else block in a while loop is executed when the condition becomes false:

```
count = 0
while count < 3:
  count += 1
  print("Hello")
else:
  print("In Else Block")
```

## Nested Loops

You can use one loop inside another loop, which is called a nested loop. Here is an example of nested for loops:

```
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]
for i in adj:
  for j in fruits:
    print(i, j)
```

## Loop Control Statements

Python supports three loop control statements: break, continue, and pass.

### Break Statement

The break statement is used to exit the loop:

```
for letter in 'science':
  if letter == 'e':
    break
  print('Current Letter :', letter)
```

### Continue Statement

The continue statement is used to skip the current iteration and continue with the next:

```
for letter in 'science':
  if letter == 'e' or letter == 's':
    continue
  print('Current Letter :', letter)
```

### Pass Statement

The pass statement is used to write empty loops or control statements:

```
for letter in 'science':
  pass
print('Current Letter :', letter)
```

# If Statements

The if statement in Python is used to execute a block of code only if a specified condition is true. It allows for conditional execution of code blocks.

## Basic If Statement

Here is an example of a simple if statement:

```
number = int(input('Enter a number: '))

if number > 0:
  print(f'{number} is a positive number.')

print('This statement is always executed.')
```

In this example, the code inside the if block executes only if the condition number > 0 is true.

## If-Else Statement

The if-else statement provides an alternative path of execution when the if condition is false.

```
number = int(input('Enter a number: '))

if number > 0:
  print('Positive number')
else:
  print('Not a positive number')

print('This statement always executes')

```
Here, if the condition number > 0 is false, the code inside the else block executes.

## If-Elif-Else Statement

The if-elif-else statement allows for multiple conditions to be checked sequentially.

```
number = int(input('Enter a number: '))

if number > 0:
  print('Positive number')
elif number < 0:
  print('Negative number')
else:
  print('Zero')

print('This statement always executes')
```
In this example, multiple conditions are checked in sequence until one is found to be true.

## Nested If Statements

You can nest if statements within each other to check multiple conditions.

```
number = int(input('Enter a number: '))

if number >= 0:
  if number == 0:
    print('Number is zero')
  else:
    print('Number is positive')
else:
  print('Number is negative')

print('This statement always executes')
```

Here, an if statement is nested inside another if statement to check additional conditions.

**Note**: Proper indentation is *crucial* in Python to define the scope of the if, else, and elif blocks

````python
number = int(input('Enter a number: '))

if number >= 0:
  if number == 0:
    print('Number is zero')
  else:
    print('Number is positive')
else:
  print('Number is negative')

print('This statement always executes')
````

## Matplotlib

Let's start by just plotting a simple sine wave between $-2\pi$ to $2\pi$.

````python
# import matplotlib and numpy (note: if already imported, this isn't necessary)
import matplotlib.pyplot as plt
import numpy as np

n = np.pi/25  # spacing increment between data points
x = np.arange(-2*np.pi, 2*np.pi + n, n)   # x-axis data points
y = np.sin(x)   # y-axis data points

plt.plot(x, y)                # creates a line plot of y vs x
plt.xlabel('Radians [rad]')   # x-axis label
plt.ylabel('Amplitude')       # y-axis label
plt.title('Sine Wave')        # plot title
plt.show()                    # command to show the plot
````

The above example is completely fine, but maybe we would like to change the x-axis units to be shown in units of $\pi$ instead of integers. We can also specify the range for both axis. However, this requires a slightly different approach to coding the figure.

````python
# import matplotlib and numpy (note: if already imported, this isn't necessary)
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, FuncFormatter

n = np.pi/25  # spacing increment between data points
x = np.arange(-2*np.pi, 2*np.pi + n, n)   # x-axis data points
y = np.sin(x)   # y-axis data points

fig, ax = plt.subplots()
ax.plot(x, y)

# Set the x-axis to increments of pi
#ax.xaxis.set_major_locator(MultipleLocator(base=np.pi))
#ax.xaxis.set_major_formatter(FuncFormatter(lambda val, pos: f'{val/np.pi:.0f}π' if val != 0 else '0'))

ax.set_xlim(-2*np.pi, 2*np.pi)  # set x-axis range
ax.set_ylim(-1, 1)              # set y-axis range
ax.set_xlabel('Radians [rad]')
ax.set_ylabel('Amplitude')
ax.set_title('Sine Wave')
plt.grid()
plt.show()
````

Clearly, both `plt.plot` and `fig, ax = plt.subplots()` can be used in Matplotlib for creating plots, but they serve different purposes and offer different levels of control.

`plt.plot`
* Usage: This is a simple and quick way to create a basic plot.
* Function: It plots data directly onto the current figure and axes.

*Example*:
```
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.show()
```
* Best for: Quick, single plots where you don't need much customization.


`fig, ax = plt.subplots()`
* Usage: This is a more flexible and powerful way to create plots.
* Function: It creates a figure (fig) and a set of subplots (ax). You can then use ax to add plots and customize them.

*Example*:
```
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```
* Best for: Complex plots, multiple subplots, and when you need more control over the layout and appearance of your plots.

**Key Differences**
* Control: plt.plot is simpler but less flexible. `fig, ax = plt.subplots()` gives you more control over the figure and axes.
* Customization: With `fig, ax = plt.subplots()`, you can easily add titles, labels, and other customizations to each subplot.
* Multiple Plots: `fig, ax = plt.subplots()` is ideal for creating multiple subplots in a single figure.

# Class Problem 3

The Barometric Equation describes the pressure as a function of altitude. It can be derived from the basic laws of physics and has the following form:

$$
\begin{align}
\frac{dP}{dz} &= -\frac{gM}{RT}P \tag{1}
\end{align}
$$

Where:
* $P$ is pressure (in Pa)
* $z$ is height (in m)
* $g = 9.81 \ (\frac{m}{s^2})$
* $M = 0.02896 \ (\frac{kg}{mol})$ is the mean molecular mass of dry air
* $R = 8.314 \ (\frac{m^3 \ Pa}{K \ mol})$ is the ideal gas constant
* $T$ is pressure (in K)

The equation can be solved for the pressure at some altitude $z$ starting at an initial pressure $P_0$ at initial height $z = 0$:

$$
\begin{align}
P = P_0e^{-\frac{z}{H}} \tag{2}
\end{align}
$$

Where we have introduced a scale factor $H = \frac{RT}{Mg} \approx 8 \ km$.

1) Write a function that will calculate equation 2 given some initial $P_0$ and height $z$.
